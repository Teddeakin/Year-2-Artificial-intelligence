```python
# Function to predict the chance of stroke based on inputted values
def chance_of_stroke(gender, age, hypertension, heart_disease, smoking):
    # Format got from chatgpt 
    
    # High risk if Female and over the age of 65 etc
    if gender == "Female" and age > 65:
        return "High risk"
    if smoking == "smokes":
        return "high risk"
    if hypertension == "1":
        return "high risk"
    if heart_disease == "1":
        return "high risk"
    
    if smoking == "formerly smoked":
        return "moderate risk"
    if age >= 45:
        return "moderate risk"
    
    # default risk level if other conditions don't match
    return "low risk"

# collecting the user inputs 
gender1 = input("What is your gender?")
# standardizing the gender input to match the function
if gender1 == "female" or gender1 == "Female":
    gender1 = "Female"
else: gender1 = "Male"

age1 = int(input("What is your current age?"))

hypertension1 = input("Do you suffer from hypertension?")
if hypertension1 == "Yes" or hypertension1 == "yes" or hypertension1 == "1" :
    hypertension1 = "1"
    hypertension2 = 1
else: 
    hypertension1 = "0" 
    hypertension2 = 0

heart_disease1 = input("Do you suffer from heart disease?")
if heart_disease1 == "Yes" or heart_disease1 == "yes" or hypertension1 == "1":
    heart_disease1 = "1"
    heart_disease2 = 1
else: 
    heart_disease1 = "0"
    heart_disease2 = 0

smoking_status1 = input("Do you currently smoke (smokes) or have been a smoker in the past (former)")
if smoking_status1 == "former":
    smoking_status1 = "formerly smoked"
elif smoking_status1 == "smokes":
    smoking_status1 = "smokes"
else: 
    smoking_status1 = "never smoked"

# calling on function to predict stroke risk based on users inputs
Test = chance_of_stroke(gender=gender1, age=age1, hypertension=hypertension1, heart_disease=heart_disease1, smoking=smoking_status1)

# displaying the stroke risk
print(f"Stroke Risk: {Test}")

# filter age over 65 from chatgpt
if age1 > 65:
    df_over_65 = df[df['age'] > 65]
else: df_over_65 = df[df['age'] < 65]

# grouping together data from the dataset based on selected values
group = df_over_65.groupby(["gender", "hypertension", "heart_disease", "smoking_status"])

# calculating the mean of stroke occuring in each group
mean = group["stroke"].mean()

# mean.loc got from chatgpt 
# finds the mean stroke chance in the users group
mean_filter = (mean.loc[(gender1, hypertension2, heart_disease2, smoking_status1)]) * 100
print("Roughly",mean_filter, "percent of people with a similar status of you have experienced a stroke")
```

```python
# function to define whether a person is at a high risk of stroke
def high_risk_of_stroke(person):

# returns true if any of following conditions match
# has to be female and over the age of 65 to return true
    if Female(person):
        if over_65:
            return True
        
    if hypertension(person):
        return True
    
    if heart_disease(person):
        return True
    
    if smoking(person):
        return True
    
    # if none of conditions of met, return false
    return False

# What each condition is trying to match
def Female(person):
    return person["gender"] == "Female"

def over_65(person):
    return person["age"] >= 65

def hypertension(person):
    return person["hypertension"] == 1

def heart_disease(person):
    return person["heart_disease"] == 1

def smoking(person):
    return person["smoking_status"] == "smokes"

# example of participant or sample data
person = {
    "gender": "Female",
    "age": 65,
    "hypertension": 1,
    "heart_disease": 1,
    "smoking_status": "smokes"
}

# runs the function with the data of "person"
comfirmed_high_risk = high_risk_of_stroke(person)
print("high risk:", comfirmed_high_risk)

# following code only runs if high risk is confirmed
if comfirmed_high_risk == True:

    # applying the function conditions to the dataset:
    # filter the dataset for the individuals that have confirment to have had a stroke 
    df_stroke = df[df["stroke"] == 1]
    
    # Create a new column in the dataset indicating whether each individual is at high risk (chatGPT)
    df["confirmed_high_risk"] = df_stroke.apply(high_risk_of_stroke, axis=1)
    
    # calculate how many people match with at least one condition
    Total_matches = df["confirmed_high_risk"].sum()
    
    # finding the percentage of those that match the high risk criteria in the dataset
    mean_high_risk = (Total_matches / 5110) * 100
    
    # prints of the calculated percentage
    print("Percentage of datasets that meet at least 1 of the requirements to have a high risk of stroke:", mean_high_risk)
```
