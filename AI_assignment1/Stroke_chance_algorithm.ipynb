```python
import pandas as pd
```

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
df.drop(["ever_married"], axis=1, inplace=True)
df.drop(["work_type"], axis=1, inplace=True)
df.drop(["Residence_type"], axis=1, inplace=True)
df.drop(["id"], axis=1, inplace=True)
df.drop(["avg_glucose_level"], axis=1, inplace=True)
df.drop(["bmi"], axis=1, inplace=True)
```

```python
df.sample(7)
```

```python
df.groupby([ "gender","age","hypertension", "heart_disease", "smoking_status"])["stroke"].agg({"mean"})
```

```python
df.groupby(["hypertension", "heart_disease"])["stroke"].agg({"mean"})
```

```python
def chance_of_stroke(gender, age, hypertension, heart_disease, smoking):
    # Format got from chatgpt 

    if gender == "female" or gender == "Female" and age > 65:
        return "High risk"
    if smoking == "smokes":
        return "high risk"
    if hypertension == "yes" or hypertension == "Yes":
        return "high risk"
    if heart_disease == "yes" or hypertension == "Yes":
        return "high risk"
    
    if smoking == "former":
        return "moderate risk"
    if age >= 45 and age <= 65:
        return "moderate risk"
    
    return "low risk"

gender1 = input("What is your gender?")
age1 = int(input("What is your current age?"))
hypertension1 = input("Do you suffer from hypertension?")
heart_disease1 = input("Do you suffer from heart disease?")
smoking_status1 = input("Do you currently smoke (smokes) or have been a smoker in the past (former)")

Test = chance_of_stroke(gender=gender1, age=age1, hypertension=hypertension1, heart_disease=heart_disease1, smoking=smoking_status1)

print(f"Stroke Risk: {Test}")
```
